```python
import os
import numpy as np
import tifffile as tiff
from stardist import export_imagej_rois
from stardist.models import StarDist2D
from csbdeep.utils import normalize
from skimage.io import imsave
import tensorflow as tf

# Disable GPU and force CPU usage
tf.config.set_visible_devices([], 'GPU')

print("Running on CPU only.")

# Define the input and output folders
input_folder = 'D:/OPS_Nov_2024/TIFF/gamma_adjusted_images'  # input folder
output_folder = 'D:/OPS_Nov_2024/TIFF/gamma_adjusted_images/Nuclear_masks'  # output folder

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Load the pretrained StarDist2D model
model = StarDist2D.from_pretrained('2D_versatile_fluo')

# Set probability and non-maximum suppression thresholds...THese are the default values but they seem to work well
prob_thresh = 0.4
nms_thresh = 0.3

# Option to save as PNG
save_as_png = True

# Function to segment an image using StarDist
def segment_image(image_path, output_path):
    # Load and normalize the image
    img = tiff.imread(image_path)
    img_normalized = normalize(img, 1, 99.8)

    # Ensure the model is trained appropriately for the DAPI images used
    labels, _ = model.predict_instances(img_normalized, prob_thresh=prob_thresh, nms_thresh=nms_thresh)

    # Save the segmented labels as either TIFF or PNG
    if save_as_png:
        output_path = output_path.replace('Phenotype_Tile', 'Phenotype_NucleiMask_Tile').replace('.tif', '.png').replace('.tiff', '.png')
        imsave(output_path, labels.astype(np.uint16))
    else:
        output_path = output_path.replace('Phenotype_Tile', 'Phenotype_NucleiMask_Tile')
        tiff.imwrite(output_path, labels.astype(np.uint16))

# Process only DAPI images in the input folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.tif', '.tiff')) and 'dapi' in filename.lower():
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, f'{filename}')
        segment_image(input_path, output_path)
        print(f'Segmented and saved: {output_path}')

print("Nuclei Segmentation completed!")
```

```python
import os
import numpy as np
import tifffile as tiff
import matplotlib.pyplot as plt
from stardist.plot import render_label

# Define the input and output folders
input_folder = 'S:/CellPose Training Images/CellPoseTraining2/TIFF/gamma_adjusted_images/Nucleus'  # Replace with your input folder
output_folder = 'S:/CellPose Training Images/CellPoseTraining2/TIFF/gamma_adjusted_images/Nucleus/masks'  # Replace with your output folder

# Function to visualize original and segmented masks side by side
def visualize_images(input_folder, output_folder):
    image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.tif', '.tiff', '.png'))]
    mask_files = [f for f in os.listdir(output_folder) if f.lower().endswith(('.tif', '.tiff', '.png'))]

    fig, axes = plt.subplots(len(image_files), 2, figsize=(12, 6 * len(image_files)))

    for i, (img_file, mask_file) in enumerate(zip(image_files, mask_files)):
        img_path = os.path.join(input_folder, img_file)
        mask_path = os.path.join(output_folder, mask_file)

        # Load the original image and mask
        original_img = tiff.imread(img_path) if img_file.lower().endswith(('.tif', '.tiff')) else plt.imread(img_path)
        mask_img = tiff.imread(mask_path) if mask_file.lower().endswith(('.tif', '.tiff')) else plt.imread(mask_path)

        # Normalize for better visualization
        image_norm = (original_img - original_img.min()) / (original_img.max() - original_img.min())

        # Count the number of detected nuclei
        num_nuclei = len(np.unique(mask_img)) - 1  # Subtract 1 to exclude background

        # Visualization with overlap
        axes[i, 0].imshow(original_img, cmap='gray')
        axes[i, 0].set_title('Original Image')
        axes[i, 0].axis('off')

        axes[i, 1].imshow(image_norm, cmap='gray')
        axes[i, 1].imshow(render_label(mask_img, img=image_norm, cmap='rainbow'), alpha=0.5)
        axes[i, 1].set_title(f'Segmented Mask with Overlay\nNuclei: {num_nuclei}')
        axes[i, 1].axis('off')

    plt.tight_layout()
    plt.show()

# Call the visualization function
visualize_images(input_folder, output_folder)
```
